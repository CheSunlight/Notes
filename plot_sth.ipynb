```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

### Known Funtion

```python
# Define function that need to plot
def func(x): 
    return 1/np.sqrt(2*np.pi) * np.exp(-x**2/2)

x = np.linspace(-4, 4, 10000) # range of x
y = func(x)
```

### Known Distribution

```python
x = np.linspace(-3, 2, 1000) # range of x
y = stats.skewnorm.pdf(x, -4, 1) # set the distribution

# delete 0's
to_delete = np.where(y==0)
x = np.delete(x, to_delete)
y = np.delete(y, to_delete)
```

### Plot

```python
fig, ax = plt.subplots(figsize=(14, 7), dpi=600)

# set the coordinate
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.plot(1, 0, '>k', transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, '^k', transform=ax.get_xaxis_transform(), clip_on=False)

ax.plot(x, y)

ax.set_xlabel('$x$', loc='right', fontsize=12)
ax.set_ylabel('$f(x)$', loc='top', fontsize=12, rotation=0)
plt.show()
```
